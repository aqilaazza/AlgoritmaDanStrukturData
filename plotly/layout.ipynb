#Modifikasi Layout

```python
import plotly.express as px

df = px.data.iris()
```

```python
df = px.data.iris()
df
```

#Memodifikasi Judul Plot

Sumber:

https://plotly.com/python/reference/layout/

```python
# Menambah judul
fig = px.scatter(df, x="sepal_length",
                 y="sepal_width",
                 color="species",
                title="Scatter Plot Sepal Width and Sepal Length")
fig.show()
```

```python
# Mengatur posisi judul
fig = px.scatter(df, x="sepal_length", y="sepal_width", color="species")

# Membuat dictionary fitur
title_dic = {
    'text': "Scatter Plot Sepal Width and Sepal Length",
    'y':0.95, # Posisi minimal y:0 dibawah dan maksimal y:1 diatas
    'x':0.5, # Posisi minimal x:0 dibawah dan maksimal x:1 diatas
}

fig.update_layout(
    title= title_dic
    )

fig.show()
```

#Dua cara berbeda mengisi parameter update_layout

```python
fig = px.scatter(df, x="sepal_length", y="sepal_width", color="species")

# CARA 1
# Membuat dictionary fitur
title_dic = {
    'text': "Scatter Plot Sepal Width and Sepal Length",
    'y':0.95, # Posisi minimal y:0 dibawah dan maksimal :1 diatas
    'x':0.5, # Posisi minimal x:0 dibawah dan maksimal x:1 diatas
}

xax_dic = {
    'title': 'Sepal Length (cm)'
}

yax_dic = {
    'title': 'Sepal Width (cm)'
}

fig.update_layout(
    title= title_dic,
    xaxis= xax_dic,
    yaxis= yax_dic
    )

fig.show()
```

#Modifikasi Traces

Dalam Plotly Express, "update traces" merujuk pada modifikasi elemen-elemen visual dari plot, seperti garis, marker, dan area. Meskipun Plotly Express adalah API tingkat tinggi yang biasanya menyembunyikan banyak detail implementasi, Anda masih dapat mengakses dan mengubah properti dari "traces" menggunakan metode update_traces.

```python
df = px.data.gapminder().query("year==2007 and continent=='Asia'")

df.sample(3)
```

- country: Nama negara, seperti "Indonesia", "India", atau "Brazil". Ini adalah kolom kategorikal yang menunjukkan negara yang sedang dianalisis.

- continent: Benua tempat negara tersebut berada, seperti "Asia", "Europe", "Africa", "Americas", atau "Oceania".

- year: Tahun dalam bentuk numerik, yang biasanya mencakup rentang dari sekitar tahun 1950-an hingga 2000-an. Ini adalah kolom waktu yang menunjukkan tahun pengukuran data.

- lifeExp (Life Expectancy): Harapan hidup rata-rata di negara tersebut pada tahun tertentu, diukur dalam tahun.

- gdpPercap (GDP per Capita): Produk domestik bruto (PDB) per kapita, atau pendapatan rata-rata per orang dalam negara tersebut dalam satu tahun. Biasanya dinyatakan dalam dolar AS.

- pop (Population): Populasi total negara pada tahun tersebut, dinyatakan dalam jumlah orang.






#Mengubah Warna Marker

```python
fig = px.scatter(df, x='gdpPercap', y='lifeExp', color='continent', title="GDP vs Life Expectancy")
fig.update_traces(marker_color='red')  # Mengubah warna marker menjadi merah
fig.show()
```

#Mengubah Ukuran Marker

```python
fig = px.scatter(df, x='gdpPercap', y='lifeExp', size='pop', color='continent', title="GDP vs Life Expectancy with Population Size")
fig.update_traces(marker_size=20)  # Mengubah ukuran marker menjadi 20
fig.show()
```

#Mengubah Bentuk Marker

```python
fig = px.scatter(df, x='gdpPercap', y='lifeExp', color='continent', title="GDP vs Life Expectancy")
fig.update_traces(marker_symbol='diamond')  # Mengubah bentuk marker menjadi diamond
fig.show()
```

#Transparansi

```python
fig = px.scatter(df, x='gdpPercap', y='lifeExp', color='continent', title="GDP vs Life Expectancy")
fig.update_traces(opacity=0.5)  # Mengatur transparansi marker menjadi 0.5
fig.show()
```

#Mengubah Posisi Text

```python
fig = px.scatter(df, x="gdpPercap", y="lifeExp", text="country", log_x=True, size_max=60)

fig.update_traces(textposition='top center')

fig.update_layout(
    height=800,
    title_text='GDP and Life Expectancy (Americas, 2007)'
)

fig.show()
```
